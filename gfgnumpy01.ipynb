List vs numpy array

```python
py_list=[1,2,3]
print(py_list * 2)
```

```python
import numpy as np
np_array=np.array([1,2,3]) #element wise multiplication
print(np_array * 2)
```

```python
import time
start = time.time()
py_list=[i*2  for i in range(10000000)]
print("list:", time.time()-start)
start=time.time()
np_array=np.array([i*2 for i in range(10000000)])
print("numpy array:",time.time()-start)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr.shape) #shape of array
print(arr.ndim) #dimension of an array
print(arr.dtype) #datatype of an array
print(arr.size)  #size of an array
```

Array reshaping

```python
arr=np.arange(12)
print("original array:" , arr)
arr1 = arr.reshape((3,4))
print("Reshaped array :", arr1)
arr2=arr1.flatten()
print("Flattened array :",arr2)
```

```python
raveled= arr1.ravel()
raveled #returns view instead of copy
#transpose
transpose = arr1.T
transpose
```
