```python
lst = []
def prime():
  for i in range(2,100):
    for j in range(2,int(i**0.5)+1):
      if i % j == 0:
        break
    else:
        lst.append(i)
prime()
# print(lst)
arr=np.array(lst)
# type(arr)
arr2=arr.reshape(len(lst),1)
arr2.ndim
# arr2
```

create a 2d numpy array of shape (3,3), filled with random integer between 1 and 10.

```python
import numpy as np
import random
data=[]
for i in range(3):
  row=[]
  for j in range(3):
    row.append(random.randint(1,10))
  data.append(row)
arr=np.array(data)
arr
#another method
arr=np.random.randint(1,11,(3,3))
arr

```
